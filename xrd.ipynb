```python
from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
# HfNiSn
mp_key = "fnNhNeZr2KXTJ64m73UhQ93prbDkSyqH"
mp_id = "mp-924128"
```

```python
with MPRester(mp_key) as m:
    
    structure = m.get_structure_by_material_id(mp_id)
    print("Téléchargement ok")
```

## Calculs

```python
#Calcule le diffractogramme
xrd = XRDCalculator('CuKa',1.54060)

#Calcule le motif de diffraction
pat = xrd.get_pattern(structure)
```

## Premier pic de diffraction

```python
print("Indices hkl associés:", pat.hkls[0])
print("2\u03F4(°) =", pat.x[0],"°")
print("Intensité =",pat.y[0])  
print("Distance interplan =",pat.d_hkls[0])
```
